```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(df["Generation"], df["Units Sales"], color='red', marker='o')
plt.xlabel("Generation")
plt.ylabel("Units Sales")
plt.title("Units Sales by Generation")
plt.tight_layout()
plt.show()
```
